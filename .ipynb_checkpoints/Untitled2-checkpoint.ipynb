```python
import numpy as np

def add(a,b):
    return f'[{a},{b}]'
def depthQ(string):
    brackets = 0
    for char in string:
        if char == '[':
            brackets += 1
        elif char == ']':
            brackets -= 1
        if brackets > 4:
            return True
    return False

import json
flatten = lambda *n: (e for a in n
    for e in (flatten(*a) if isinstance(a, (tuple, list)) else (a,)))
def splitQ(string):
    if any( int(char)>=10 for char in flatten(json.loads(string))  ): 
        return True
    return False 

def reduce(string):
    do, result = True, string
    while do: 
        do = False
        if depthQ(result):
            do, result = True, explode2(result)
        elif splitQ(result):
            do, result = True, split(result)
    return result


def split(string):
    result = ''
    numlist = flatten(json.loads(string))
    value = ''
    for num in numlist:
        if num >= 10:
            value = str(num)
            break
    start = string.find(value)
    end = start + len(value)
    left = int(value) // 2
    right= int(float(value) / 2 + 0.5)
    result = string[:start] + "[" + str(left) + ","  + str(right) + "]" + string[end:]
    return result

def explode(string):
    brackets = 0
    result = ''
    for i, char in enumerate(string):
        if char == '[':
            brackets += 1
        elif char == ']':
            brackets -= 1
        if brackets > 4 and string[i+1]!='[':
            start = i
            end = string.find(']',i)
            left,right = list(filter(None,re.split("\[|,|\]",string[start:end+1])))
          #  left, right = ll[0],ll[1]
            
            if re.findall("[0-9]",string[:start]):
                searching = True
                LEDigit = -1
                LSDigit = -1
                loc = start
                while searching:
                    loc -= 1
                    if string[loc] >= '0' and string[loc] <= '9':
                        if(LEDigit < 0):
                            #found end of some number
                            LEDigit = loc
                        if string[loc - 1] < '0' or string[loc - 1] > '9':
                            searching = False
                            LSDigit = loc
                newLeft = int(string[LSDigit:LEDigit + 1]) + int(left)
                result += string[0:LSDigit] + str(newLeft) + string[LEDigit + 1:start] + '0'
                
            else: 
                result += string[:start]+'0'
                
            if re.findall("[0-9]",string[end+1:]):
                searching = True
                REDigit = -1
                RSDigit = -1
                loc = end+1
                while searching:
                    loc += 1
                    if string[loc] >= '0' and string[loc] <= '9':
                        if(REDigit < 0):
                            #found end of some number
                            REDigit = loc
                        if string[loc + 1] < '0' or string[loc + 1] > '9':
                            searching = False
                            RSDigit = loc
                newRight = int(string[RSDigit:REDigit + 1]) + int(right)
                result += string[end+1:RSDigit] + str(newRight) + string[REDigit + 1:]     
            else: 
                result += string[end+1:]
            break
    return result

def explode2(number: str):
    result = ""
    bracket = 0
    i = 0
    while True:
        if number[i] == '[':
            bracket += 1
        elif number[i] == ']':
            bracket -= 1
        if bracket > 4 and number[i+1] != '[': #Boom Boom!
            #get both numbers
            start = i
            end = number.find("]", start)
            tokens = list(filter(None,re.split("\[|,|\]",number[start:end+1])))

            left = int(tokens[0])
            right = int(tokens[1])

            #add left leftwards if possible
            if re.findall("[0-9]",number[0:start]):
                    #find first number, and add it
                    searching = True
                    LEDigit = -1
                    LSDigit = -1
                    loc = start
                    while searching:
                        loc -= 1
                        if number[loc] >= '0' and number[loc] <= '9':
                            if(LEDigit < 0):
                                #found end of some number
                                LEDigit = loc
                            if number[loc - 1] < '0' or number[loc - 1] > '9':
                                searching = False
                                LSDigit = loc
                    newLeft = int(number[LSDigit:LEDigit + 1]) + left
                    result += number[0:LSDigit] + str(newLeft) + number[LEDigit + 1:start] + '0' #put current spot as 0 as well!
            else:
                result += number[0:start] + '0'
                        
            #add right rightwards if possible
            if re.findall("[0-9]",number[end+1:]):
                    #find first number, and add it
                    searching = True
                    REDigit = -1
                    RSDigit = -1
                    loc = end + 1
                    while searching:
                        loc += 1
                        if number[loc] >= '0' and number[loc] <= '9':
                            if(RSDigit < 0):
                                #found start of some number
                                RSDigit = loc
                            if number[loc + 1] < '0' or number[loc + 1] > '9':
                                searching = False
                                REDigit = loc
                    newRight = int(number[RSDigit:REDigit + 1]) + right
                    result += number[end+1:RSDigit] + str(newRight) + number[REDigit + 1:]
            else:
                result += number[end+1:]
            break
        i += 1
    return result

def mag(crab):
    if type(crab) == int:
        return crab
    if type(crab) == list:
        left, right = crab
        return 3*mag(left) + 2*mag(right)
    
def magsym(crab):
    if type(crab) == int:
        return crab
    if type(crab) == list:
        left, right = crab
        return magsym(left) + magsym(right)

testnum1 = '[[[[4,3],4],4],[7,[[8,4],9]]]'
testnum2 = '[1,1]'
import json, re
#print(reduce(add(testnum1,testnum2)))
#print(reduce(add(testnum1,testnum2)))
#print(mag(json.loads('[[[[8,7],[7,7]],[[8,6],[7,7]]],[[[0,7],[6,6]],[8,7]]]')))

inlist = open('input_full.txt').read().splitlines()

total = inlist[0]
for line in inlist[1:]:
    total = add(total,line.strip())
    total = str(reduce(total))
print(mag(json.loads(total)))
    
```

```python
def invmerge(A):
    if len(A)==1:
        return A,0
    else:
        mid = len(A)//2
        L, invL = invmerge(A[:mid])
        R, invR = invmerge(A[mid:])
        sortA = []
        inv = invL + invR
        i, j = 0,0
        while i < len(L) and j < len(R):
            if L[i] < R [j]:
                sortA.append(L[i])
                i += 1
            else:
                sortA.append(R[j])
                j += 1
                inv += (len(L)-i)
        sortA += L[i:]
        sortA += R[j:]
        return sortA, inv
```

```python

def pidgeonhole_sort(L,infimum, supremum):
    counting = [L.count(i) for i in range(infimum, supremum+1)]
    srtlist = []
    for index, freq in enumerate(counting):
        for _ in range(freq):
            srtlist.append(index)
    return srtlist
pidgeonhole_sort([3,2,4,5,6,6,6,7,1],0,8)
```

```python
class node:
    def __init__(self, data = None):
        self.data = data
        self.next = None
class linkedlist:
    def __init__(self):
        self.head = node()
    def append(self,data):
        newnode = node(data)
        cur = self.head
        while cur.next != None:
            cur = cur.next
        cur.next = newnode
    def length(self):
        cur = self.head
        total = 0
        while cur.next != None:
            cur = cur.next
            total += 1
        return total
    def display(self):
        cur = self.head
        elems = []
        while cur.next != None:
            cur = cur.next
            elems.append(cur.data)
        print(elems)
```

```python
def insertion_sort(A):
    for j in range(1,len(A)):
        key = A[j]
        i = j - 1
        while i >= 0 and A[i] > key:
            A[i+1] = A[i]
            i -= 1
        A[i+1] = key
    return A

def selection_sort(A):
    j = 0
    while j < len(A):
        min = A[j]
        mindex = j
        for i in range(j,len(A)):
            if A[i] < min:
                min = A[i]
                mindex = i
        A[mindex] = A[j]
        A[j] = min
        j += 1
    return A

def noninc_insertion_sort(A):
    for j in range(1,len(A)):
        key = A[j]
        i = j - 1
        while i >= 0 and A[i] < key:
            A[i+1] = A[i]
            i -= 1
        A[i+1] = key
    return A

def merge(A,p,q,r):
    n1 = q - p + 1
    n2 = r - q 
    L = [0]*(n1+1)
    R = [0]*(n2+1)
    for i in range(n1):
        L[i]=A[p+i]
    for i in range(n2):
        R[i]=A[q+1+i]
    L[n1] = float("inf")
    R[n2] = float("inf")
    i = 0
    j = 0
    for k in range(p,r+1):
        if L[i]<R[j]:
            A[k]=L[i]
            i += 1
        else:
            A[k]=R[j]
            j += 1
    return A
def merge_sort(A,p = 0 ,r = len(A)):
    if p < r:
        q = ( p + r ) // 2
        merge_sort(A,p,q)
        merge_sort(A,q+1,r)
        merge(A,p,q,r)
    return A
    
def rec_insertion_sort(A):
    n = len(A)
    if n > 1:
        rec_insertion_sort( A[:-1] )
        i = n-2
        key = A[-1]
        while i >= 0 and A[i] > key:
            A[i+1] = A[i]
            i -= 1
        A[i+1] = key
    return A

def iter_binary_search(A,nu):
    high = len(A)-1
    low = 0
    while high >= low:
        mid = ( high + low)// 2
        if A[mid] == nu:
            return mid
        elif A[mid] > nu:
            high = mid - 1
        else:
            low = mid + 1
    return -1

def iter_binary_search(A,nu):
    high = len(A)-1
    low = 0
    while high >= low:
        mid = ( high + low)// 2
        if A[mid] == nu:
            return mid
        elif A[mid] > nu:
            high = mid - 1
        else:
            low = mid + 1
    return -1

def rec_binary_search(A,nu, low = 0, high = len(A) - 1):
    if high < low: return -1
    mid = ( high + low)// 2
    if A[mid] == nu:
        return mid
    elif A[mid] > nu:
        return rec_binary_search(A,nu,low,mid - 1)
    else:
        return rec_binary_search(A,nu,mid + 1,high)
        
```
