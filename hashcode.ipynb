```python
# integer partitions by https://jeromekelleher.net/generating-integer-partitions.html
def partitions(n):
    a = [0 for i in range(n + 1)]
    k = 1
    y = n - 1
    while k != 0:
        x = a[k - 1] + 1
        k -= 1
        while 2 * x <= y:
            a[k] = x
            y -= x
            k += 1
        l = k + 1
        while x <= y:
            a[k] = x
            a[l] = y
            yield a[:k + 2]
            x += 1
            y -= 1
        a[k] = x + y
        y = x + y - 1
        yield a[:k + 1] 
        
def digit_sum(string):
    return sum(map(int,[*string]))

from itertools import cycle, permutations, repeat

def combination_insert(insert_list,target):
    l1 = [*target]
    l2 = [*insert_list]

    combined_indices = range(len(l1)+len(l2))
    i1next = cycle(l1).__next__

    l3 = [[pos_to_l2[pos] if pos in pos_to_l2 else i1next() for pos in combined_indices]
        for pos_to_l2 in map(dict, map(zip, permutations(combined_indices, len(l2)), repeat(l2)))]

    return [ ''.join(c for c in string) for string in l3]
    
```
